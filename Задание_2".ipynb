Подгружаем библиотеку пандас

```python
import pandas as pd
```

Считываем файл excel

```python
df = pd.read_excel('/content/Задание по консолидации данных_.xlsx', sheet_name='данные', header=2)
type(df)
```

```python
years = [
        pd.read_excel('/content/Задание по консолидации данных_.xlsx', sheet_name='данные', header=1).columns[2],
        pd.read_excel('/content/Задание по консолидации данных_.xlsx', sheet_name='данные', header=1).columns[2+12]
        ]
years
```

```python
df1=df.loc[:,"Магазин":"Город"]
df1["Город"]=df1["Город"].fillna(method = 'ffill')
df1
```

```python
df2=df.loc[:,"Январь":'Декабрь']
df2 = df2.add_suffix(' ' + str(years[0]))
df2 = df2.fillna(0)
df2
```

```python
df3=df.loc[:,"Январь.1":]
df3.columns = df3.columns.str.replace('.1', ' ')
df3=df3.add_suffix(' '+ str(years[1]))
df3=df3.fillna(0)
df3
```

```python
ndf = pd.concat([df1,df2,df3], axis =1)
ndf
```

```python
table = pd.melt(ndf,id_vars = ['Магазин','Город'], value_vars = ndf.loc[:], var_name="Дата", value_name = 'Прибыль,руб')
table
```

```python
table['Дата'].str.split(' ', expand=True)
```

```python
table['Месяц'] = table['Дата'].str.split(' ', expand=True)[0]
table['Месяц']
```

```python
table['Год'] = table['Дата'].str.split(' ', expand=True)[1]
table['Год']
```

```python
table
```

```python
month = pd.DataFrame(table['Месяц'].unique()).reset_index()
month.columns = ['№ Месяца', 'Месяц']
month['№ Месяца'] = month['№ Месяца']+1
month
```

```python
year = table['Год'].unique()
year
```

```python
table = table.merge(month, left_on ="Месяц",right_on ="Месяц" )
table
```

```python
q = [
    (1,1),
    (2,1),
    (3,1),
    (4,2),
    (5,2),
    (6,2),
    (7,3),
    (8,3),
    (9,3),
    (10,4),
    (11,4),
    (12,4)
]
```

```python
q = pd.DataFrame(q, columns = ['№ Месяца', '№ Квартала'])
q
```

```python
table = table.merge(q, left_on ="№ Месяца",right_on ="№ Месяца" )
table
```

```python
res = pd.pivot_table(table[table['№ Квартала']==1],
                    index = 'Город',
                    columns = 'Год',
                    values = "Прибыль,руб",
                    aggfunc = 'sum', #сумма всех магазинов для конкретного города
                    )
res
```

```python
res.style.background_gradient(cmap = 'Blues')
```

```python
res['Изменение в %'] = ((res['']-res['2021'])/res['2021'])*100
res
```
